# **Title**
Exploratory Data Analysis of Amazon Prime Video Content Library**

# **Objective**
To analyze Amazon Prime Video’s content library using Exploratory Data Analysis (EDA)
techniques and identify trends in content type, genres, release years, and ratings.

```python
#Load Libraries & Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
titles = pd.read_csv('titles[1].csv')
credits = pd.read_csv('credits[1].csv')

# Preview data
titles.head()
```

```python
#Dataset Information
titles.info()

```

```python
#Statistical Summary
titles.describe()
```

```python
#Movie vs TV Show Count
plt.figure(figsize=(6,4))
titles['type'].value_counts().plot(kind='bar')
plt.title('Movie vs TV Show Count on Amazon Prime')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
#Genres Count
# Split genres and count
genres = titles['genres'].dropna().str.split(',').explode()

plt.figure(figsize=(8,5))
genres.value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Genres on Amazon Prime')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
#Release Year Trend
plt.figure(figsize=(8,5))
titles['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Content Release Trend Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

# **Conclusion**
The EDA reveals that Amazon Prime Video hosts more movies than TV shows, with a steady
increase in content production over recent years. Drama and Comedy emerge as dominant
genres, indicating audience preference trends on the platform.
